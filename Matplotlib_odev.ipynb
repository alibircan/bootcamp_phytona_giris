```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline

n_groups = 8

dataF = pd.DataFrame()
dataF['menscore'] = np.random.rand(n_groups)*100
dataF['womenscore'] = np.random.rand(n_groups)*100
dataF.index= 2019-n_groups +np.arange(n_groups)

bar_width = 0.8

plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
rects1 = plt.bar(dataF.index, dataF['menscore'], bar_width,color='b',label='Men')
plt.xlabel('Group')
plt.ylabel('Scores')
plt.title('Men Scores by years')

plt.subplot(1, 2, 2)
rects2 = plt.bar(dataF.index , dataF['womenscore'],  bar_width, color='r', label='Women')
plt.xlabel('Group')
plt.ylabel('Scores')
plt.title('Women Scores by years')
plt.show()
```
